```python
# Step 3: Test the data fetcher
import sys
sys.path.insert(0, '.')

from src.kanyakumari_monitor import KanyakumariOceanMonitor

monitor = KanyakumariOceanMonitor()
print("üîÑ Fetching real-time data...")

marine = monitor.fetch_marine_data()
weather = monitor.fetch_weather_data()
earthquakes = monitor.fetch_earthquakes()

print(f"\nüåä Wave Height: {marine.get('wave_height', 'N/A')} m")
print(f"üå°Ô∏è Temperature: {weather.get('temperature', 'N/A')}¬∞C")
print(f"üåç Earthquakes: {len(earthquakes)} found")
print("\n‚úÖ Data fetcher working!")
```

```python
# Step 4: Start the FastAPI server
import subprocess
import threading
import time

def run_server():
    subprocess.run(['python', '-m', 'uvicorn', 'src.kanyakumari_api:app', '--host', '0.0.0.0', '--port', '8000'])

server_thread = threading.Thread(target=run_server, daemon=True)
server_thread.start()

time.sleep(3)
print("‚úÖ FastAPI server started on port 8000")
```

```python
# Step 5: Test the API
import requests

try:
    response = requests.get('http://127.0.0.1:8000/api/status', timeout=10)
    print("API Response:")
    print(response.json())
    print("\n‚úÖ API is working!")
except Exception as e:
    print(f"‚ùå Error: {e}")
```
